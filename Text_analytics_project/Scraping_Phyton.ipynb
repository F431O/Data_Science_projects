```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import ElementClickInterceptedException
```

```python
def start_driver():
    ser = Service("./chromedriver")
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument(r'user-data-dir=C:\Users\PC\AppData\Local\Google\Chrome\User Data') # QUI
    chrome_options.add_argument(r'--profile-directory=Default')
    driver = webdriver.Chrome(
    service=ser,
    options=chrome_options)
    return driver
```

```python
data = pd.read_csv("Restaurant.csv")
```

```python
text = list()
ID = list()
date = list()
user = list()
rate = list()
```

```python
driver=start_driver()
#urls = data["Url"][520:620] #il primo numro mettere l'indice in cui si Ã¨ fermato meno uno
urls = data["Url"][692:720]

for URL in urls:
    
    rist = data[data["Url"]==URL]["ID"].iloc[0]
    
    driver.get(URL)

    next_page=True
    print("ID ",rist)
    start_page=1


    for i in range(5):
        time.sleep(10)
        try:
            checkbox = driver.find_elements(By.CLASS_NAME, "ui_checkbox")
            checkbox[i].click()
        except:
            next
        
        time.sleep(5)
        
        t=0      
        while next_page and t<4:
        
            print(start_page, end=", ")
            clik = True
            time.sleep(10)
        
            button_next = driver.find_elements(By.CLASS_NAME, "filters_detail_checkbox_trating_cbx")

            extend_text = driver.find_elements(By.CSS_SELECTOR, "DIV.prw_rup.prw_reviews_text_summary_hsx .partial_entry span.taLnk")
            if extend_text != [] and clik:
                extend_text[0].click()
                clik = False
                time.sleep(10)

            html = driver.page_source
            soup = BeautifulSoup(html)
            rl = soup.find("div", {"class": "ratings_and_types block_wrap ui_section"})
            rlbox = rl.find_all("div", {"class": "ui_column is-9"})
            userbox = rl.find_all("div", {"class": "ui_column is-2 memberInfoColumn"})
        
            for r in userbox:
                try:
                    text_completo = r.find("div",{"class":"info_text pointer_cursor"}).text
                    da_eliminare = r.find("div",{"class":"userLoc"})
                    if da_eliminare is not None:
                        delete = da_eliminare.text
                        username = text_completo.replace(delete,"")
                    else:
                        username = text_completo
                except:
                    username = None
                
                user.append(username)
                


            for r in rlbox:
            
                quattro = r.find("span", {"class":"ui_bubble_rating bubble_40"})
                tre = r.find("span", {"class":"ui_bubble_rating bubble_30"})
                due = r.find("span", {"class":"ui_bubble_rating bubble_20"})
                uno = r.find("span", {"class":"ui_bubble_rating bubble_10"})
                cinque = r.find("span", {"class":"ui_bubble_rating bubble_50"})
            
                if uno is not None:
                    rate.append(1)
                if due is not None:
                    rate.append(2)
                if tre is not None:
                    rate.append(3)
                if quattro is not None:
                    rate.append(4)
                if cinque is not None:
                    rate.append(5)
                
                text.append(r.find("p", {"class": "partial_entry"}).text)
                date.append(r.find("span", {"class":"ratingDate"})["title"])
                ID.append(rist)
      
            time.sleep(5)
            button_next = driver.find_elements(By.CLASS_NAME, "next")
            try:
                button_next[0].click()
                start_page+=1
                t+=1
            except:
                next_page=False
                t=5
                


        time.sleep(10)
        try:
            checkbox = driver.find_elements(By.CLASS_NAME, "ui_checkbox")
            checkbox[i].click()
        except:
            next
    



dict_reviews = {"ID":ID,"Rating":rate,"Review_Date":date,"User":user,"Review":text}
Reviews = pd.DataFrame(dict_reviews)
Reviews.to_csv("Reviews_600_700.csv",index=False)        
```
